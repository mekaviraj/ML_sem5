```python
# Load data
df = pd.read_csv("house_price_train.csv")
X = df[['OverallQual']].values
y = df['SalePrice'].values

# Normalize
X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
m, n = X.shape[0], X.shape[1]
X = np.c_[np.ones((m, 1)), X] # add bias

# Hypothesis & Cost
def predict(X, theta): return X.dot(theta)
def cost(X, y, theta): return 1/(2*m)*np.sum((predict(X, theta)-y)**2) #mean()

# Gradient Descent
def gradient_descent(X, y, alpha, epochs):
    theta = np.zeros(X.shape[1])
    history = []
    for i in range(epochs):
        theta = theta - alpha/m * X.T.dot(predict(X, theta) - y)
        history.append(cost(X, y, theta))
    return theta, history

# Try different learning rates
theta_1, history_1 = gradient_descent(X, y, 0.01, 100)
theta_2, history_2 = gradient_descent(X, y, 0.2, 100)
theta_3, history_3 = gradient_descent(X, y, 0.5, 100)
plt.plot(history_1, label='a=0.01')
plt.plot(history_2, label='a=0.2')
plt.plot(history_3, label='a=0.5')
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Cost over time")
plt.legend()
plt.show()
```

## comparing desicison tree vs random forest


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load and preprocess dataset
data = load_breast_cancer()
X, y = data.data, data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Initialize classifiers
dt_classifier = DecisionTreeClassifier(random_state=42)
rf_classifier = RandomForestClassifier(random_state=42, n_estimators=100)

# Train and evaluate Decision Tree
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)

# Train and evaluate Random Forest
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
```

```python
# Calculate metrics
def calculate_metrics(y_true, y_pred, model_name):
    return {
        'Model': model_name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, average='weighted'),
        'Recall': recall_score(y_true, y_pred, average='weighted'),
        'F1-Score': f1_score(y_true, y_pred, average='weighted')
    }

dt_metrics = calculate_metrics(y_test, y_pred_dt, "Decision Tree")
rf_metrics = calculate_metrics(y_test, y_pred_rf, "Random Forest")

# Display results
results_df = pd.DataFrame([dt_metrics, rf_metrics])
print("\nPerformance Comparison:")
print(results_df.round(4))
```

```python

# Effect of number of estimators on Random Forest
n_estimators_range = [1, 5, 10, 25, 50, 75, 100, 150, 200]
accuracy_scores = []

for n_est in n_estimators_range:
    rf_temp = RandomForestClassifier(n_estimators=n_est, random_state=42)
    cv_scores = cross_val_score(rf_temp, X, y, cv=5, scoring='accuracy')
    accuracy_scores.append(cv_scores.mean())
```

```python
# Plot results
plt.figure(figsize=(10, 5))
plt.plot(n_estimators_range, accuracy_scores, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Estimators')
plt.ylabel('Cross-Validation Accuracy')
plt.title('Accuracy vs Number of Estimators in Random Forest')
plt.grid(True, alpha=0.3)
plt.show()

# Summary
print("\n=== Summary of Results ===")
print(f"Decision Tree Accuracy: {dt_metrics['Accuracy']:.4f}")
print(f"Random Forest Accuracy: {rf_metrics['Accuracy']:.4f}")
print(f"Improvement from DT to RF: {(rf_metrics['Accuracy'] - dt_metrics['Accuracy']):.4f}")
```

## comparing Max Voting, Average Voting, and Weighted Average Voting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score

# Load dataset
wine_data = load_wine()
X_wine, y_wine = wine_data.data, wine_data.target

# Preprocessing
from sklearn.preprocessing import StandardScaler
scaler_wine = StandardScaler()
X_wine_scaled = scaler_wine.fit_transform(X_wine)

```

```python

# Split data
X_train_w, X_test_w, y_train_w, y_test_w = train_test_split(X_wine_scaled, y_wine, test_size=0.3, random_state=42, stratify=y_wine)

# Define base classifiers
classifiers = [
    ('lr', LogisticRegression(random_state=42, max_iter=1000)),
    ('svm', SVC(random_state=42, probability=True)),
    ('knn', KNeighborsClassifier()),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42, n_estimators=50))
]

```

```python
# Individual classifier performance
individual_results = []
for name, clf in classifiers:
    clf.fit(X_train_w, y_train_w)
    y_pred = clf.predict(X_test_w)
    accuracy = accuracy_score(y_test_w, y_pred)
    individual_results.append((name, accuracy))

```

```python

# Ensemble methods
# 1. Max Voting (Hard Voting)
voting_hard = VotingClassifier(estimators=classifiers, voting='hard')
voting_hard.fit(X_train_w, y_train_w)
y_pred_hard = voting_hard.predict(X_test_w)
hard_accuracy = accuracy_score(y_test_w, y_pred_hard)

# 2. Average Voting (Soft Voting)
voting_soft = VotingClassifier(estimators=classifiers, voting='soft')
voting_soft.fit(X_train_w, y_train_w)
y_pred_soft = voting_soft.predict(X_test_w)
soft_accuracy = accuracy_score(y_test_w, y_pred_soft)

# 3. Weighted Average Voting
weights = [0.15, 0.20, 0.15, 0.25, 0.25]  # Manually assigned weights based on model performance
voting_weighted = VotingClassifier(estimators=classifiers, voting='soft', weights=weights)
voting_weighted.fit(X_train_w, y_train_w)
y_pred_weighted = voting_weighted.predict(X_test_w)
weighted_accuracy = accuracy_score(y_test_w, y_pred_weighted)
```

```python

# Display results
ensemble_results = pd.DataFrame({
    'Method': ['Max Voting', 'Average Voting', 'Weighted Average'],
    'Accuracy': [hard_accuracy, soft_accuracy, weighted_accuracy]
})
print(ensemble_results.round(4))

# Compare with best individual classifier
best_individual = max(individual_results, key=lambda x: x[1])
print(f"\nBest Individual Classifier: {best_individual[0]} - {best_individual[1]:.4f}")
print(f"Best Ensemble Method: Weighted Average - {weighted_accuracy:.4f}")
```

```python

# Visualize ensemble performance
methods = [name.upper() for name, _ in individual_results] + ['MAX', 'AVG', 'WGT']
accuracies = [acc for _, acc in individual_results] + [hard_accuracy, soft_accuracy, weighted_accuracy]

colors = ['blue'] * len(individual_results) + ['red'] * 3

plt.bar(methods, accuracies, color=colors, alpha=0.7)
plt.axhline(y=best_individual[1], color='green', linestyle='--', label=f'Best Individual ({best_individual[0]})')
plt.xlabel('Classification Methods')
plt.ylabel('Accuracy')
plt.title('Individual vs Ensemble Classifier Performance')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

print("=== TASK 3: Hard Voting vs Soft Voting ===")

# Load dataset
data = load_iris()
X, y = data.data, data.target
```

```python

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Preprocessing - Simple scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# Define base classifiers
dt = DecisionTreeClassifier(random_state=42)
lr = LogisticRegression(random_state=42, max_iter=1000)
knn = KNeighborsClassifier()
svm = SVC(probability=True, random_state=42)  # probability=True for soft voting

# Individual classifier performance
classifiers = [('Decision Tree', dt), ('Logistic Regression', lr),
               ('KNN', knn), ('SVM', svm)]

print("\nIndividual Classifier Performance:")
for name, clf in classifiers:
    if name in ['KNN', 'SVM']:
        clf.fit(X_train_scaled, y_train)
        y_pred = clf.predict(X_test_scaled)
    else:
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name}: {acc:.4f}")
```

```python

# Hard Voting Ensemble
hard_voting = VotingClassifier(
    estimators=[('dt', dt), ('lr', lr), ('knn', knn), ('svm', svm)],
    voting='hard'
)
hard_voting.fit(X_train, y_train)
y_pred_hard = hard_voting.predict(X_test)
hard_accuracy = accuracy_score(y_test, y_pred_hard)

# Soft Voting Ensemble
soft_voting = VotingClassifier(
    estimators=[('dt', dt), ('lr', lr), ('knn', knn), ('svm', svm)],
    voting='soft'
)
soft_voting.fit(X_train, y_train)
y_pred_soft = soft_voting.predict(X_test)
soft_accuracy = accuracy_score(y_test, y_pred_soft)

print(f"\nHard Voting Accuracy: {hard_accuracy:.4f}")
print(f"Soft Voting Accuracy: {soft_accuracy:.4f}")

# Show probability outputs for soft voting
probabilities = soft_voting.predict_proba(X_test[:5])
print(f"\nSoft Voting Probabilities (first 5 samples):")
print(probabilities)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

print("\n=== TASK 5: RandomForestRegressor with OOB Score ===")

# Load regression dataset
data = load_diabetes()
X, y = data.data, data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

# Experiment with different numbers of trees
n_trees_list = [10, 50, 100, 200, 500]
oob_scores = []
test_scores = []

print("OOB Scores vs Test Scores:")
for n_trees in n_trees_list:
    rf_reg = RandomForestRegressor(
        n_estimators=n_trees,
        oob_score=True,
        random_state=42
    )
    rf_reg.fit(X_train, y_train)

    # Predictions
    y_pred = rf_reg.predict(X_test)

    # Scores
    oob_score = rf_reg.oob_score_
    test_r2 = r2_score(y_test, y_pred)
    test_mse = mean_squared_error(y_test, y_pred)

    oob_scores.append(oob_score)
    test_scores.append(test_r2)

    print(f"Trees: {n_trees:3d} | OOB Score: {oob_score:.4f} | Test R: {test_r2:.4f} | Test MSE: {test_mse:.4f}")

```

```python
# Compare OOB score with cross-validation
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(RandomForestRegressor(n_estimators=100), X, y, cv=5, scoring='r2')
print(f"\nCross-Validation R: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
```

```python

# Plot OOB vs Test scores
plt.figure(figsize=(10, 6))
plt.plot(n_trees_list, oob_scores, 'bo-', label='OOB Score')
plt.plot(n_trees_list, test_scores, 'ro-', label='Test R Score')
plt.xlabel('Number of Trees')
plt.ylabel('Score')
plt.title('OOB Score vs Test R Score')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Final model with best parameters
final_rf = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=42)
final_rf.fit(X_train, y_train)

print(f"\nFinal Model:")
print(f"OOB Score: {final_rf.oob_score_:.4f}")
print(f"Test R: {r2_score(y_test, final_rf.predict(X_test)):.4f}")
print(f"Number of features: {X.shape[1]}")
print(f"OOB prediction shape: {final_rf.oob_prediction_.shape}")
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate dataset
X, y_true = make_blobs(n_samples=200, centers=3, cluster_std=0.60, random_state=42)

# Function: Divisive Clustering (DIANA Simulation using KMeans splits)
def divisive_clustering(X, k=3):
    clusters = [X]   # start with all data as one cluster

    while len(clusters) < k:
        # Pick the largest cluster to split
        largest_cluster = max(clusters, key=len)
        clusters.remove(largest_cluster)

        # Split into 2 using KMeans
        km = KMeans(n_clusters=2, random_state=42).fit(largest_cluster)
        y = km.labels_

        # Create new clusters
        clusters.append(largest_cluster[y==0])
        clusters.append(largest_cluster[y==1])

    return clusters

# Run DIANA simulation
clusters = divisive_clustering(X, k=3)

# Plot results
colors = ['red', 'green', 'blue']
for i, cluster in enumerate(clusters):
    plt.scatter(cluster[:,0], cluster[:,1], color=colors[i])
plt.title("Divisive Clustering (DIANA Simulation)")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.metrics import jaccard_score


iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print("Sample Dataset:\n", df.head())


p1 = df.iloc[0].to_numpy()
p2 = df.iloc[1].to_numpy()

def euclidean_distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))

def manhattan_distance(p1, p2):
    return np.sum(np.abs(p1 - p2))

def cosine_similarity(p1, p2):
    dot_product = np.dot(p1, p2)
    return dot_product / (np.linalg.norm(p1) * np.linalg.norm(p2))

def minkowski_distance(p1, p2, p):
    return np.power(np.sum(np.abs(p1 - p2) ** p), 1/p)



binary_df = (df > df.mean()).astype(int)
v1 = binary_df.iloc[0].to_numpy()
v2 = binary_df.iloc[1].to_numpy()


print("\nBetween row 0 and row 1 in Iris dataset:")
print("Euclidean:", euclidean_distance(p1, p2))
print("Manhattan:", manhattan_distance(p1, p2))
print("Cosine Similarity:", cosine_similarity(p1, p2))
print("Minkowski (p=3):", minkowski_distance(p1, p2, 3))
print("Jaccard Index:", jaccard_score(v1, v2))

```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster


X, y = make_blobs(n_samples=100, centers=3, cluster_std=1.2, random_state=42)

linkage_methods = ['single', 'complete', 'average', 'ward', 'centroid']

for method in linkage_methods:
    linked = linkage(X, method=method)
    plt.figure(figsize=(8, 4))
    dendrogram(linked)
    plt.title(f"Hierarchical Clustering - {method.capitalize()} Linkage")
    plt.xlabel("Samples")
    plt.ylabel("Distance")
    plt.show()

    # Form clusters (optional)
    labels = fcluster(linked, t=3, criterion='maxclust')
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
    plt.title(f"{method.capitalize()} Linkage Clustering Result")
    plt.show()
```

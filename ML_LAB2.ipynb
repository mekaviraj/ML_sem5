```python
import pandas as pd
from sklearn.datasets import load_diabetes
from pandas import DataFrame as df
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
# Load data(diabetes)
data=load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
# prep('target' label)
df['target'] = df['target'].apply(lambda x: df['target'].mean() < x).astype(int)
df['target_class'] = df['target'].apply(lambda x: df['target'].mean() < x).astype(int)
X = df.drop(['target', 'target_class'], axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3, random_state=2)
clf=DecisionTreeClassifier(max_depth=4, random_state=2)
clf.fit(X_train, y_train)
y_predict=clf.predict(X_test)
print(f"Accuracy: {metrics.accuracy_score(y_test, y_predict)}")
```

```python
# Load data
df = pd.read_csv("house_price_train.csv")
X = df[['OverallQual']].values
y = df['SalePrice'].values

# Normalize
X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
m, n = X.shape[0], X.shape[1]
X = np.c_[np.ones((m, 1)), X] # add bias

# Hypothesis & Cost
def predict(X, theta): return X.dot(theta)
def cost(X, y, theta): return 1/(2*m)*np.sum((predict(X, theta)-y)**2) #mean()

# Gradient Descent
def gradient_descent(X, y, alpha, epochs):
    theta = np.zeros(X.shape[1])
    history = []
    for i in range(epochs):
        theta = theta - alpha/m * X.T.dot(predict(X, theta) - y)
        history.append(cost(X, y, theta))
    return theta, history

# Try different learning rates
theta_1, history_1 = gradient_descent(X, y, 0.01, 100)
theta_2, history_2 = gradient_descent(X, y, 0.2, 100)
theta_3, history_3 = gradient_descent(X, y, 0.5, 100)
plt.plot(history_1, label='a=0.01')
plt.plot(history_2, label='a=0.2')
plt.plot(history_3, label='a=0.5')
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Cost over time")
plt.legend()
plt.show()
```
